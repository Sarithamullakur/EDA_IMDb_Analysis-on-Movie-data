# **6. Data Pre-Processing**
---

- This section is emphasised on performing data manipulation over unstructured data for further processing and analysis.

- To modify unstructured data to strucuted data you need to verify and manipulate the integrity of the data by:
  - Handling missing data,

  - Handling redundant data,

  - Handling inconsistent data,

  - Handling outliers,

  - Handling typos

```python
'''
1. Missing data in Revenue (Millions) column - 128 missing values
2. Missing data in Metascore - 64 missing values
3. Runtime (Minutes) - Data type is in int64. But this should be continuous variable
4. Total Titles are 999 instead of 1000. Check for the duplicate Title
'''
```

```python
Mov_data.columns
```

```python
# 1. Missing data in Metascore column - 64 missing values
# Will replace the missing items with Mean
Mov_data['Metascore'] = Mov_data['Metascore'].fillna(Mov_data['Metascore'].mean())
```

```python
# 2. Missing data in Revenue (Millions) column - 128 missing values. Fill with Median
Mov_data['Revenue (Millions)'] = Mov_data['Revenue (Millions)'].fillna(Mov_data['Revenue (Millions)'].median())
```

```python
#Verify
Mov_data.info()
```

```python
# Total Titles are 999 instead of 1000. Check for the duplicate Title
Mov_data.shape[0] - Mov_data['Title'].nunique()
```

```python
Mov_data['Title'].value_counts()
```

The Host movie is displayed twice, However the directors are distinct and year of relase is 2006 and 2013. 

```python
#4. Runtime (Minutes) - Data type is in int64. But this should be continuous variable
Mov_data['Runtime (Minutes)'] = Mov_data['Runtime (Minutes)'].astype('float64')
```

```python
# Year is in int64 format. This should be in dateformat
# And split year from the datetime
Mov_data['Year'] = pd.to_datetime(Mov_data['Year'],format='%Y')
Mov_data['yr'] = Mov_data['Year'].dt.year
Mov_data['yr']
```

#### 8.12 List of Actors who acted in highly generated Revenue

```python
Mov_data_actors = Mov_data.assign(Actors = Mov_data['Actors'].str.split(",") )
Mov_data_actors = Mov_data_actors.explode('Actors')
```

```python
#Creating a custom figure size of 15*7 inches
paper = plt.figure(figsize = [15,7])

data = Mov_data_actors.groupby(['Actors']).aggregate(['min','max','mean'])['Revenue (Millions)'].sort_values(by=['mean'],ascending=False)[:15]

data.plot.bar()

#Labelling x-axis with a custom label and size of 14
plt.xlabel('Actors', size = 14, color = 'blue')

#Labelling y-axis with a custom label and size of 14
plt.ylabel('Revenue', size = 14, color = 'blue')

#Labelling title with a custom label and size of 16
plt.title('Revenue and Actors Relation', size = 16)

# Display the output by rendering visual on the screen
plt.show()
```

```python
Mov_data_actors['Actors'] = Mov_data_actors['Actors'].astype(str).str.lstrip().str.rstrip().str.title()
Mov_data_actors
```

```python
#Creating a custom figure size of 15*7 inches
paper = plt.figure(figsize = [15,7])

data=(Mov_data_actors[Mov_data_actors['Revenue (Millions)'] > 300].value_counts('Actors')/Mov_data_actors['Actors'].value_counts()[Mov_data_actors['Actors'].value_counts() > 2]).sort_values(ascending=False)[:10]
data.plot.bar()

#Labelling x-axis with a custom label and size of 14
plt.xlabel('Actors', size = 14, color = 'blue')

#Labelling y-axis with a custom label and size of 14
plt.ylabel('% of Movies Generated 300M+ Revenue', size = 14, color = 'blue')

#Labelling title with a custom label and size of 16
plt.title('Actors Who Were Part of 300M+ Movies More Than Twice', size = 16)

# Display the output by rendering visual on the screen
plt.show()
```

- In Top 10 **Revenue** generation movies **no actors are repeated**
- **Oscar Isac** seems to be part of 1 **top Revenue** generation movies. But the **high and low revenue** difference doesnt imply that he could be the sole reason for high revenue generation
- Gwyneth Paltow and Tyrse Gibson were succesful in more than 2 movies that generated revenue greater than 300+ millions.
